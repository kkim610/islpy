# What is Statistical Learning?
Example: We are statistical consultant hired to help improve **sales** based on sales data from 200 different markets with advertising budgets for **TV**, **radio**, and **newspaper**. The customer can only control advertising expenditures, not sales directly. Our job is to uncover relationships between advertising and sales and leverage that information to build a model to predict sales based on the three media budgets.

*input variables* are typically denoted as $X$ with a subscript to distinguish them. In this example $X_1$ might be **TV**, $X_2$ might be **radio**, and $X_3$ might be **newspaper**. Also known as *predictors, independent variables, features, variables*
  
*output variables* are typically denoted as $Y$ also known as *response, dependent variable*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
advert = pd.read_csv('../data/Advertising.csv', index_col=0)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 5))

sns.regplot(x='TV', y='sales', data=advert, ax=ax1, scatter_kws={'color': 'red'}, 
            line_kws={'color': 'blue'}, ci=None)

sns.regplot(x='radio', y='sales', data=advert, ax=ax2, scatter_kws={'color': 'red'}, 
            line_kws={'color': 'blue'}, ci=None)

sns.regplot(x='newspaper', y='sales', data=advert, ax=ax3, scatter_kws={'color': 'red'}, 
            line_kws={'color': 'blue'}, ci=None)
```

More generally than this example, suppose we want to observe a quantitative response $Y$ from $p$ different predictors, $X_1, X_2,...,X_p$. Assuming there is some relationshipt between $Y$ and $X=(X_1, X_2,...,X_p)$, it can be written as:
  
$$Y = f(X) + \epsilon$$

where $f$ is a fixed but unknown function and $\epsilon$ is a random *error term* which is independent of $X$ and has mean zero. $f$ represents the *systematic* information that $X$ provides about $Y$.

Statistical learning refers to a set of approaches for estimating $f$

## Why estimate $f$

#### Prediction
In many cases $X$ is readily available but $Y$ cannot be easily obtained. Since $\epsilon$ averages to zero, we can predict $Y$ with $$\widehat{Y} = \widehat{f}(X),$$ where $\widehat{f}$ is our estimate for $f$ and $\widehat{Y}$ is the resulting prediction for $Y$. Here $\widehat{f} is often treated as a *black box*.

The accuracy of $\widehat{Y}$ as a prediction for $Y$ depends on *reducible error* and *irreducible error*. In general, $\widehat{f}$ will not be a perfect estimate for $f$, and this will introduce some error. This is called *reducible error* because we can improve the accuracy of $\widehat{f}$ by using the most appropriate statistical learning technique to estimate $f$. However, even if it were possible to create a perfect estimate for $f$ (e.g. $\widehat{Y} = f(X)$) it would still have error in it because $\widehat{Y}$ is a function of $\epsilon$ which by definition cannot be predicted with $X$. This is known as *irreducible error*.

Irrecudible error, $\epsilon$, is larger than zero because:  
* the quantity may contain unmeasured variables that are useful to predict $Y$
* the quantity may contain unmeasured variation

Consider the prediction $\widehat{Y} = \widehat{f}(X)$ and assume that $\widehat{f}$ and $X$ are fixed. It's easy to show that
$$E(Y - \widehat{Y})^2 = E[f(X) + \epsilon - \widehat{f}(X)]^2$$
$$ = [f(X) - \widehat{f}(X)]^2 + Var(\epsilon),$$ where $E(Y - \widehat{Y})^2$ represents the average, or *expected value*, of the squared difference between the predicted and actual value of $Y$, and $Var(\epsilon)$ represents the *variance* associated with the error term $\epsilon$. This book focuses on estimating $f$ with the aim of minimizing reducible error.

#### Inference
Often we want to understand the way that $Y$ is affected by $X$. In this situation we want to estimate $f$ but not necessarily $Y$. Now $\widehat{f}$ cannot be treated as a black box, because we need to know it's exact form. We may be interested in answering the following questions:
* *Which predictors are associated with the response?*
* *What is the relationship between the response and each predictor?*
* *Can the relationship between $Y$ and each predictor be adequately summarized using a linear equation, or is the relationship more complicated?*

Considering the advertising example earlier, we may be interested in answering questions such as:
* *Which media contribute to sales?*
* *Which media generate the biggest boost in sales?*
* *How much increase in sales is associated with a given increase in TV advertisint?*

This falls into the inference paradigm. 

Often, modeling will be conducted for both prediction and inference. For example, in real estate, one may be interested in how individual input variables affect the price (e.g. crime rate, zoning, distance from river, air quality, etc.), which would be inference. On the other hand, one may simply want to predict the sale price of a home based on the input variables, which would be prediction.

*Linear models* give simple and interpretable inference, but may not yield accurate results. In contrast, some highly non-linear approaches can provide quite accurate results for predicting $Y$, but will be less interpretable.

## How do we estimate $f$
In this book we will always assume the following:
* we have observed $n$ data points which are called the *training data*
* let $x_{ij}$ represent the $j$th predictor for observation $i$
    * $i = 1, 2, ..., n$
    * $j = 1, 2, ..., p$
* let $y_i$ represent the response variable for the $i$th observation
* then our training data consists of $\{(x_1, y_1), (x_2, y_2), ..., (x_n, y_n)\}$
    * where $x_i = (x_{i1}, x_{i2}, ..., x_{ip})^T$
    
Our goal is to find a function $\widehat{f}$ such that $Y \approx \widehat{f}(X)$ for any observation $(X, Y)$
