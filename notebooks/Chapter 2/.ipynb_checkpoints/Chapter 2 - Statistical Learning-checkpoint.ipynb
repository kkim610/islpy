# What is Statistical Learning?
Example: We are statistical consultant hired to help improve **sales** based on sales data from 200 different markets with advertising budgets for **TV**, **radio**, and **newspaper**. The customer can only control advertising expenditures, not sales directly. Our job is to uncover relationships between advertising and sales and leverage that information to build a model to predict sales based on the three media budgets.

*input variables* are typically denoted as $X$ with a subscript to distinguish them. In this example $X_1$ might be **TV**, $X_2$ might be **radio**, and $X_3$ might be **newspaper**. Also known as *predictors, independent variables, features, variables*
  
*output variables* are typically denoted as $Y$ also known as *response, dependent variable*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
advert = pd.read_csv('../../data/Advertising.csv', index_col=0)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 5))

sns.regplot(x='TV', y='sales', data=advert, ax=ax1, scatter_kws={'color': 'red'}, 
            line_kws={'color': 'blue'}, ci=None)

sns.regplot(x='radio', y='sales', data=advert, ax=ax2, scatter_kws={'color': 'red'}, 
            line_kws={'color': 'blue'}, ci=None)

sns.regplot(x='newspaper', y='sales', data=advert, ax=ax3, scatter_kws={'color': 'red'}, 
            line_kws={'color': 'blue'}, ci=None)
```

More generally than this example, suppose we want to observe a quantitative response $Y$ from $p$ different predictors, $X_1, X_2,...,X_p$. Assuming there is some relationshipt between $Y$ and $X=(X_1, X_2,...,X_p)$, it can be written as:
  
$$Y = f(X) + \epsilon$$

where $f$ is a fixed but unknown function and $\epsilon$ is a random *error term* which is independent of $X$ and has mean zero. $f$ represents the *systematic* information that $X$ provides about $Y$.

Statistical learning refers to a set of approaches for estimating $f$

## Why estimate $f$

#### Prediction
In many cases $X$ is readily available but $Y$ cannot be easily obtained. Since $\epsilon$ averages to zero, we can predict $Y$ with  

$$\widehat{Y} = \widehat{f}(X),$$  
where $\widehat{f}$ is our estimate for $f$ and $\widehat{Y}$ is the resulting prediction for $Y$. Here $\widehat{f}$ is often treated as a *black box*.

The accuracy of $\widehat{Y}$ as a prediction for $Y$ depends on *reducible error* and *irreducible error*. In general, $\widehat{f}$ will not be a perfect estimate for $f$, and this will introduce some error. This is called *reducible error* because we can improve the accuracy of $\widehat{f}$ by using the most appropriate statistical learning technique to estimate $f$. However, even if it were possible to create a perfect estimate for $f$ (e.g. $\widehat{Y} = f(X)$) it would still have error in it because $\widehat{Y}$ is a function of $\epsilon$ which by definition cannot be predicted with $X$. This is known as *irreducible error*.

Irrecudible error, $\epsilon$, is larger than zero because:  
* the quantity may contain unmeasured variables that are useful to predict $Y$
* the quantity may contain unmeasured variation

Consider the prediction $\widehat{Y} = \widehat{f}(X)$ and assume that $\widehat{f}$ and $X$ are fixed. It's easy to show that
$$E(Y - \widehat{Y})^2 = E[f(X) + \epsilon - \widehat{f}(X)]^2$$
$$ = [f(X) - \widehat{f}(X)]^2 + Var(\epsilon),$$ where $E(Y - \widehat{Y})^2$ represents the average, or *expected value*, of the squared difference between the predicted and actual value of $Y$, and $Var(\epsilon)$ represents the *variance* associated with the error term $\epsilon$. This book focuses on estimating $f$ with the aim of minimizing reducible error.

#### Inference
Often we want to understand the way that $Y$ is affected by $X$. In this situation we want to estimate $f$ but not necessarily $Y$. Now $\widehat{f}$ cannot be treated as a black box, because we need to know it's exact form. We may be interested in answering the following questions:
* *Which predictors are associated with the response?*
* *What is the relationship between the response and each predictor?*
* *Can the relationship between $Y$ and each predictor be adequately summarized using a linear equation, or is the relationship more complicated?*

Considering the advertising example earlier, we may be interested in answering questions such as:
* *Which media contribute to sales?*
* *Which media generate the biggest boost in sales?*
* *How much increase in sales is associated with a given increase in TV advertisint?*

This falls into the inference paradigm. 

Often, modeling will be conducted for both prediction and inference. For example, in real estate, one may be interested in how individual input variables affect the price (e.g. crime rate, zoning, distance from river, air quality, etc.), which would be inference. On the other hand, one may simply want to predict the sale price of a home based on the input variables, which would be prediction.

*Linear models* give simple and interpretable inference, but may not yield accurate results. In contrast, some highly non-linear approaches can provide quite accurate results for predicting $Y$, but will be less interpretable.

## How do we estimate $f$
In this book we will always assume the following:
* we have observed $n$ data points which are called the *training data*
* let $x_{ij}$ represent the $j$th predictor for observation $i$
    * $i = 1, 2, ..., n$
    * $j = 1, 2, ..., p$
* let $y_i$ represent the response variable for the $i$th observation
* then our training data consists of $\{(x_1, y_1), (x_2, y_2), ..., (x_n, y_n)\}$
    * where $x_i = (x_{i1}, x_{i2}, ..., x_{ip})^T$
    
Our goal is to find a function $\widehat{f}$ such that $Y \approx \widehat{f}(X)$ for any observation $(X, Y)$

#### Parametric Models
1. Make an assumption about the functional form of $f$. For linear assumption: 
$$f(X) = \beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_pX_p$$ 
The problem of estimating $f$ is greatly simplified because we only need to estimate the $p + 1$ coefficients $\beta_0, \beta_1, ..., \beta_p$
<br>
2. Now, need a procedure that uses the training data to *fit* or *train* the model. In this case we need to estimate the parameters $\beta_0, \beta_1, ..., \beta_p$ to find the values of the parameters such that:
$$Y \approx \beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_pX_p$$
Most common approach for linear problem is *(ordinary) least squares*

#### Non-Parametric Models
Non-parametric methods make no assumption about the functional form of $f$. Instead they seek an estimate of $f$ that gets as close to the training data as possible without being too rough or wiggly. They have a potential to fit a wide range of shapes for $f$. But since they don't reduce the problem of estimating $f$ to a small number of parameters, a very large number of observations is required.

## The Trade-Off Between Prediction Accuracy and Model Interpretability
If accurate prediction is the goal without regard to model interpretability, a more flexible model would be appropriate. If understanding the relationship between inputs and outputs is the goal (e.g. inference), then we might prefer a more restrictive (less flexible) model whose coefficients easier to interpret. 

## Supervised Versus Unsupervised Learning
In an unsupervised situation, we lack a response variable that can supervise our analysis. With unsupervised learning we can seek to understand relationships between variables or observations or perform *cluster anlysis*. 

A *semi-supervised learning* problem might arise where out of $n$ observations, $m$ have a response while $n - m$ have no response.

## Regression Versus Classification Problems
Depends on whether the response is quantitative (regression) or qualitative (classification). 

# Assessing Model Accuracy
Selecting the best learning approach can be one of the most challenging parts of statistical learning in practice. This section covers some of the most important concepts that arise in selecting a statistical learning procedure.

## Measuring Quality of Fit
We need to quantify the extent to which the predicted response for a given observation is close to the true response for the same observation. In regression, the most commonly used measure is *mean squared error* (MSE):
$$MSE = \frac{1}{n}\sum_{i=1}^n(y_i - \widehat{f}(x_i))^2$$
This is the *training MSE* which will be small if the predicted responses are close to the true responses. 

However, we are interested in the accuracy of the predictions on previously unseen data (e.g. *test data*). Mathematically, we are not interested in whether $\widehat{f}(x_i) \approx y_i$, instead we want to know whether $\widehat{f}(x_0) \approx y_0$ for some test data. We want to choose the method that gives the lowest *test MSE*. With a large number of observations, we could compute:  

$$Ave(y_0 - \widehat{f}(x_0))^2,$$  
which is the average squared prediction error for the test observations $(x_0, y_0)$.

There is a tradeoff between MSE and model flexibility which will change depending on the complexity of the relationship between inputs and responses which are being modeled. Training MSE will always go down with increased model flexibility, but test MSE will usually decrease, reach a local minimum, and then increase again. Since we want to minimize the test MSE, we would choose the appropriate model flexibility based on minimum test MSE (see figures 2.9 - 2.11).

## The Bias-Variance Trade-Off
The U-shaped curve in the test MSE is the result of two competing properties of statistical learning methods, variance and bias. With the following mathematical decomposition, it can be shown that the expected test MSE can be cecomposed into the sum of three fundamental quantities: the *variance* of $\widehat{f}(x_0)$, the squared *bias* of $\widehat{f}(x_0)$, and the variance of the error terms $\epsilon$:  

$$\mathbb{E}[(y_0 - \widehat{y}_0)^2] = \mathbb{E}[y_0^2 - 2y_0\widehat{y}_0 + \widehat{y}_0^2]$$  

$$= \mathbb{E}[y_0]^2 - 2\mathbb{E}[y_0\widehat{y}_0] + \mathbb{E}[\widehat{y}_0^2]$$  

Apply this Lemma to first and third term: $\mathbb{E}[x] = \mathbb{E}[(x - \mathbb{E}[x])^2] + \mathbb{E}[x]^2$  

$$= \mathbb{E}[(y_0 - \mathbb{E}[y_0])^2] + \mathbb{E}[y_0]^2 - 2\mathbb{E}[y_0\widehat{y}_0] + \mathbb{E}[(\widehat{y}_0 - \mathbb{E}[\widehat{y}_0])^2] + \mathbb{E}[\widehat{y}_0]^2$$  

Use the fact that $\mathbb{E}[\widehat{y}_0] = \widehat{f}(x_0)$  

$$= \mathbb{E}[(y_0 - \mathbb{E}[y_0])^2] + \mathbb{E}[y_0]^2 - 2\mathbb{E}[y_0]\widehat{f}(x_0) + \mathbb{E}[(\widehat{y}_0 - \widehat{f}(x_0))^2] + \widehat{f}(x_0)^2$$  

Rearranging this equation we get  

$$= \mathbb{E}(y_0 - \mathbb{E}[y_0])^2] + \mathbb{E}[y_0]^2 - 2\mathbb{E}[y_0]\widehat{f}(x_0) + \widehat{f}(x_0)^2 + \mathbb{E}[(\widehat{y} - \widehat{f}(x_0))^2]$$  

Which can be further simplified by factorization to  

$$= \mathbb{E}[(y_0 - \mathbb{E}[y_0])^2] + (\mathbb{E}[y_0] - \widehat{f}(x_0))^2 + \mathbb{E}[(\widehat{y}_0 - \widehat{f}(x_0))^2]$$  

Since $\widehat{y}_0 - \widehat{f}(x_0) = \epsilon$ we can rewrite the third term  

$$= \color{blue}{\mathbb{E}[(y_0 - \mathbb{E}[y_0])^2} + \color{green}{(\mathbb{E}[y_0] - \widehat{f}(x_0))^2} + \color{red}{\mathbb{E}[\epsilon^2]}$$  

And since $\widehat{y}_0 = \widehat{f}(x_0) + \epsilon$ and $\mathbb{E}[\epsilon] = 0$, this is exactly the same as  

$$\mathbb{E}[(y_0 - \widehat{f}(x_0))^2]= \color{blue}{Var(\widehat{f}(x_0))} + \color{green}{[Bias(\widehat{f}(x_0))]^2} + \color{red}{Var(\epsilon)}$$  

In order to minimize the expected test error, we need to select a statistical learning method that simultaneously achieves *low variance* and *low bias*. Since variance and squared bias are nonnegative, expected test MSE can never lie below $Var(\epsilon)$  

*Variance* refers to the amount by which $\widehat{f}$ would change if we estimated it using a different training set. Ideally, the estimate for $f$ should not vary much between training sets, but if a method has high variance, then small changes in the training data can result in large changes of $\widehat{f}$. In general, more flexible models have higher variance.  

*Bias* refers to the error that is introduced by approximating a real-life problem, which may be extremely complicated, by a much simpler model. In general, more flexible models result in less bias.  

Good test set performance of a statistical method requires low variance as well as low squared bias.

## The Classification Setting
The concepts thus far have been focused on the regression setting, but many of the concepts transfer over to the classification setting with some modification due to the fact that $y_i$ is no longer numerical, but qualitative. The most common approach fro quantifying the accuracy of our estimate of $\widehat{f}$ is the training *error rate* (the proportion of mistakes that are made if we apply our estimate $\widehat{f}$ to the training observations):  

$$\frac{1}{n}\sum_{i=1}^nI(y_i \neq \widehat{y}_i)$$  
where $\widehat{y}_i$ is the predicted class label for the $i$th observation using $\widehat{f}$. $I(y_i \neq \widehat{y}_i)$ is an *indicator variable* that equals 1 if $y_i \neq \widehat{y}_i$ and zero if $y_i = \widehat{y}_i$.  

The *test error* rate associated with a set of test observations of the form $(x_0, y_0)$ is given by:  

$$Ave(I(y_0 \neq \widehat{y}_0))$$  

### The Bayes Classifier
It can be proven that the test error rate is minimized, on average, by assigning each observation to the most likely class given its predictor values. We should simply assign a test observation with predictor vector $x_0$ to the class $j$ for which  

$$Pr(Y = j | X = x_0)$$  

is largest. This is the *conditional probability* and is also known as the *Bayes classifier*. The boundary in which the probability of $Y$ is exactly equal to 50% is called the *Bayes decision boundary* (see figure 2.13).  

The bayes classifier produces the lowest possible test error rate, called the *Bayes error rate*. Since the Bayes classifier will always choose the class for which the conditional probabily is largest, the error rate at $X = x_0$ will be $1 - max_jPr(Y = j | X = x_0$). In general, the overall Bayes error rate is given by:  

$$1-\mathbb{E}[max_jPr(Y = j | X)]$$  

where the expectation averages the probability over all possible values of $X$.

### K-Nearest Neighbors
For real data, computing the Bayes classifier is impossible because the conditional distribution of $Y$ given $X$ is unknown, thus it is the unattainable gold standard against which to compare other methods. Many approaches attempt to estimate the conditional distribution of $Y$ given $X$, and then classify an observation to the class with the highest *estimated* probability.  

One such method is the *K-Nearest Neighbors* (KNN) classifier. Given a positive integer $K$ and a test observation $x_0$, the KNN classifier first identifies the $K$ points in the training data that are closest to $x_0$ (euclidean distance), represented by $N_0$. It then estimates the condional probability for class $j$ as the fraction of points in $N_0$ whose response values equal $j$:  

$$Pr(Y = j | X = x_0) = \frac{1}{K}\sum_{i \in N_0}I(y_i = j)$$

Finally, KNN applies the Bayes rule and classifies the test observation $x_0$ to the class with the largest probability. The choice of $K$ has a drastic effect on the KNN classifier obtained. $K = 1$ will be overly flexible (high variance, low bias) while large values of $K$ will be less flexible and produce a decision boundary that is close to linear (low variance, high bias).
