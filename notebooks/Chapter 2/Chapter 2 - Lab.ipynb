## Graphics

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

#this will make the plots inline
%matplotlib inline

x = np.random.normal(size=100)
y = np.random.normal(size=100)
#the semicolon at the end inhibits output
plt.scatter(x, y);
plt.xlabel('this is the x-axis');
plt.ylabel('this is the y-axis');
plt.title('Plot of X vs Y');
#plt.savefig('Figure.pdf')
#plt.savefig('Figure.jpeg')
```

```python
#calculate the delta based on the number of points wanted
delta = 2 * np.pi / 50
#python is exclusive on the top end, so need to add one more step
x = np.arange(-np.pi, np.pi + delta, delta)
y = x
#need to create a mesh from x and y
X, Y = np.meshgrid(x, y)
f = np.cos(Y) / (1 + X**2)

plt.figure()
plt.contour(X, Y, f, cmap='jet');

plt.figure()
plt.contour(X, Y, f, 45, cmap='jet');

fa = (f - f.T) / 2
plt.figure()
plt.contour(X, Y, fa, 15, cmap='jet');
```

```python
plt.figure()
plt.pcolor(X, Y, fa, cmap='jet')

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, fa, cmap='jet')
```
