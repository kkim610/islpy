combining the two and reindexing the summations we get

Since $x_i$ is a constant relative to the summation over $j$, it can be brought inside the sum. Thus, we can rewrite this as

where

**6**:  
The simple linear regression equation is  

$$y = \widehat{\beta}_0 + \widehat{\beta}_1x$$  

From (3.4)  

$$\widehat{\beta}_0 = \bar{y} - \widehat{\beta}_1\bar{x}$$  

Substituting this into the simple linear regression we get  

$$y = \bar{y} - \widehat{\beta}_1\bar{x} + \widehat{\beta}_1x$$  

The right-hand side of this equation will equal zero if the point $(\bar{x}, \bar{y})$ is on the line. We can check this by substituting $\bar{x}$ for $x$ in the above equation and checking that $y = \bar{y}$

$$y = \bar{y}$$  

Thus the simple linear regression line always passes through the point $(\bar{x}, \bar{y})$

**7**:  
$$R^2 = 1 - \frac{RSS}{TSS}$$  

where  

$$RSS = \sum_{i=1}^n(y_i - \widehat{y}_i)^2\qquad\text{and}\qquad TSS = \sum_{i=1}^n(y_i - \bar{y})^2$$  

Assuming $\bar{x} = \bar{y} = 0$ (and consequently $\widehat{\beta}_0 = \bar{y} - \widehat{\beta}_1\bar{x} = 0$) we can rewrite $R^2$ as  

$$R^2 = 1 - \frac{\sum_{i=1}^n(y_i - \widehat{y}_i)^2}{\sum_{i=1}^ny_i^2}$$  

Remembering that $\widehat{y}_i = \widehat{\beta}_1x_i$ (since $\widehat{\beta}_0 = 0)$, this equation becomes  

$$R^2 = 1 - \frac{\sum_{i=1}^n(y_i - \widehat{\beta}_1x_i)^2}{\sum_{i=1}^ny_i^2}$$    

$$=\frac{\sum_{j=1}^ny_j^2}{\sum_{j=1}^ny_j^2} - \frac{\sum_{i=1}^n(y_i^2 - 2x_iy_i\widehat{\beta}_1 + x_i^2\widehat{\beta}_1^2)}{\sum_{j=1}^ny_j^2}$$  

$$=\frac{\sum_{j=1}^ny_j^2 - \sum_{i=1}^ny_i^2 + 2\widehat{\beta}_1\sum_{i=1}^nx_iy_i - \widehat{\beta}_1^2\sum_{i=1}^nx_i^2}{\sum_{j=1}^ny_j^2}$$

Also, since $\bar{x} = \bar{y} = 0$ we can write $\widehat{\beta}_1$ as  

$$\widehat{\beta}_1 = \frac{\sum_{i=1}^nx_iy_i}{\sum_{i=1}^nx_i^2}$$  

Substuting this into the equation and cancelling the first two terms in the numerator we get  

$$R^2 = \frac{2\frac{\sum_{i=1}^nx_iy_i}{\sum_{i=1}^nx_i^2}\sum_{i=1}^nx_iy_i - \bigl(\frac{\sum_{i=1}^nx_iy_i}{\sum_{i=1}^nx_i^2}\bigr)^2\sum_{i=1}^nx_i^2}{\sum_{i=1}^ny_i^2}$$  

$$= \frac{2\frac{(\sum_{i=1}^nx_iy_i)^2}{\sum_{x=1}^nx_i^2} - \frac{(\sum_{x=1}^nx_iy_i)^2}{\sum_{x=1}^nx_i^2}}{\sum_{i=1}^ny_i^2}$$  

$$\boxed{R^2 = \frac{(\sum_{i=1}^nx_iy_i)^2}{\sum_{i=1}^nx_i^2\sum_{i=1}^ny_i^2}}$$  

From equation 3.18, the correlation is defined as

$$Cor(X, Y) = \frac{\sum_{i=1}^n(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^n(x_i - \bar{x})^2}\sqrt{\sum_{i=1}^n(y_i - \bar{y})^2}}$$  

Assuming $\bar{x} = \bar{y} = 0$, the squared correlation reduces to  

$$Cor(X, Y)^2 = \frac{(\sum_{i=1}^nx_iy_i)^2}{\sum_{i=1}^nx_i^2\sum_{i=1}^ny_i^2}$$  

Which is exactly the form we got for $R^2$ in the boxed equation above

**8**:

```python
auto = pd.read_csv('../../data/Auto.csv', na_values='?')
auto_ols = smf.ols('mpg ~ horsepower', auto).fit()
auto_ols.summary()
```

**8 (a) i**:  
We can test the null hypothesis that all regression coefficients are equal to zero, i.e. $H_0: \beta_i = 0\space \forall \space i$. The p-value corresponding to the F-statistic is 7.03e-81 which clearly indicates a relationship betwen mpg and horsepower. Thus, we can reject the null hypothesis.  

**8 (a) ii**:  

```python
mpg_pred = auto_ols.predict()
n = len(auto.dropna())
RSS = ((auto.dropna()['mpg'].values - mpg_pred)**2).sum()
RSE = np.sqrt((1/(n-2))*RSS)
mpg_mean = auto['mpg'].mean()
print('Percentage Error: {:.2f}%'.format(RSE / mpg_mean * 100))
```

To calculate the error relative to the response (mpg), we use the mean of the response and the RSE. The calculations above provide a percentage error of 20.86%. We can also see that the $R^2$ value of 0.606 indicates that almost 61% of the variance in mpg can be explained with horsepower.  

**8 (a) iii**:  
The relationship between mpg and horsepower is negative because the coefficient associated with horsepower in the ols summary above is negative. This means that increasing the horsepower by one unit will decrease the mpg by 0.1578 units.
